# Containers

- Install dependencies, configuration can be CHALLENGING and prone to errors if done manually when moving the app from Dev A to Dev B to QA
- Way to package applications with all the necessary dependencies and configuration
- It is a portable artifact (easily share and move this package to any environment)
- Makes development and deployment more easy and efficient.
- You pack and the other developer unpacks

## Virtual machines
- Both perform virtualization
- Hardware - OS Kernel - App Layer (VM will install OS layer as well, docker image only the app layer)
- We can install VM in any OS, Docker may have compatibility issues (for example for older versions of windows).

## Docker image
- In the container we create multiple layer of images. Ex one Linux layer, one mongodb image etc.
- The collection of images is called a docker image.
- Package or the artifact, can be moved or shared
- Running docker image will start the app and in order to run the app it will create a container
- They are very small
- They start and run faster than VMs


### Docker installation

- Search for the system requirements Hyper V must be enabled

### Basic commands
- In Dockerhub there are multiple public images that you can pull

** First we pull the image:

docker pull docker/getting-started:pwd

** Run the image:

docker run -d -p 80:80 docker/getting-started
Multiple images can run in the same port. Local port should be different: different:same

-d: detach mode to contiue writing commands in cmd

-p: define the port 80:80 hostcode:containercode

** To check if the app is running:

docker ps

** To show and delete the images:

docker images

docker image rm IMAGE_ID

** To stop a container:

docker stop CONTAINER_ID

### Build Image


docker build -t dockerimagename .

docker run -p  5000:5000 dockerimagename

it will generate 2 ips, one local and one ip inside the container

### Publish to ducker hub

docker login

Rename the docker image:

docker tag dockerimagename dockerimagenewname

docker push dockerimagename:latest (or version number)

## Docker compose

Tool for defining and running multi container docker applications.
Ex: One mysql container, one redis container etc
